```python
def func(df):
    df['Date'] = pd.to_datetime(df['Date'])
    '''I can't find any nan value in the Dataset
     ,but I will make function to do this :)'''
    df.fillna(df.mean(numeric_only=True), inplace=True)
    '''There isn't any type of dataset must be convert to String
     Or Category because most of dataset Attribute (string or int
      , float) and three is on of this Attributes with a type
       object and No of logical to Convert Date to String or
        Category but I will write the code to do this , Just for
         Cerfication :)'''
    df['Data_String'] = df['Date'].astype(str)

func(df)
```

```python
x = df[' Price per unit'] * df['Quantity']
print("Total Sales =", x.sum())
print("Mean =", x.mean().sum())
print("Median =", x.median().sum())
print("Max =", x.max())
print("Min =", x.min())
```

```python
df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month_name()
df
```

```python
x = df[' Price per unit'] * df['Quantity']
Top_month_sales = df.groupby(x) ['Month'].value_counts(ascending = False).head(1)
Low_month_sales = df.groupby(x) ['Month'].value_counts(ascending = True).head(1)
print('Top Month Sales', Top_month_sales)
print('Low Month Sales', Low_month_sales)
```

```python
df.columns = df.columns.str.strip()
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Total Sales'] = df['Price per unit'] * df['Quantity']
```

```python
fig_hist = px.histogram(df, x='Total Sales', title='Distribution of Total Sales', color_discrete_sequence=['red'])
fig_hist.show()
```

```python
df.columns = df.columns.str.strip()
df['Total_Sales'] = df['Price per unit'] * df['Quantity']
fig = px.line(df, x=df.index, y='Total_Sales', title="Total Sales Over Time")
fig.show()
# from this plot we conclude that trend is increase slowly (seasonaly)
```

```python
monthly_sales = df.groupby('Month')['Total_Sales'].sum()

fig_trend = px.line(monthly_sales,
                    x=monthly_sales.index,
                    y=monthly_sales.values,
                    title='Monthly Sales Trend (Total Sales per Month)',
                    markers=True,
                    color_discrete_sequence=['red']) # استخدام اللون الأحمر المفضل لديك

fig_trend.update_layout(xaxis_title='Month', yaxis_title='Total Sales')
fig_trend.show()
```

```python
product_sales = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False)
fig_prod = px.bar(product_sales, x=product_sales.index, y=product_sales.values, title='Total Sales per Product', color_discrete_sequence=['blue'])
fig_prod.show()
```

```python
avg_price = df.groupby('Product')['Price per unit'].mean()
fig_price = px.bar(avg_price, x=avg_price.index, y=avg_price.values,  title='Average Price per Product', color_discrete_sequence=['green'])
fig_price.show()
```

```python
product_sales = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False)
fig_prod = px.bar(product_sales, x=product_sales.index, y=product_sales.values,title='Total Sales per Product', color_discrete_sequence=['red'])
fig_prod.show()
```

```python
avg_price = df.groupby('Product')['Price per unit'].mean()
fig_price = px.bar(avg_price, x=avg_price.index, y=avg_price.values,title='Average Price per Product', color_discrete_sequence=['Purple'])
fig_price.show()
```

```python
manager_sales = df.groupby('Manager')['Total_Sales'].sum().sort_values(ascending=False)
fig_manager = px.bar(manager_sales, x=manager_sales.index, y=manager_sales.values,title='Manager Ranking by Sales', color_discrete_sequence=['red'])
fig_manager.show()
```

```python
city_sales = df.groupby('City')['Total_Sales'].sum().sort_values(ascending=False)
fig_city = px.pie(city_sales, names=city_sales.index, values=city_sales.values,title='Sales Distribution by City')
fig_city.show()
```

```python
purchase_compare = df.groupby('Purchase Type')['Total_Sales'].sum()
fig_purchase = px.pie(purchase_compare, names=purchase_compare.index, values=purchase_compare.values,title='Online vs In-store vs Drive-thru')
fig_purchase.show()
```

```python
payment_counts = df['Payment Method'].value_counts()
fig_payment = px.bar(payment_counts, x=payment_counts.index, y=payment_counts.values,title='Most Used Payment Methods', color_discrete_sequence=['Turquoise'])
fig_payment.show()
```
